# Explatory Data Analysis

## Hypothesis 1

Assumption: GDP per capita positively affects use of cab.

- We need to determine gdp per of states.
- Sum of transation IDs per city. (Cab table is enough)

Firstly, determine unique cities from Cab table.

```python
cities = pd.DataFrame(cab_df["city"].unique())
cities.head()
```

For GDP per capita information, I got data from Statista https://www.statista.com/statistics/248112/per-capita-us-real-gross-domestic-product-gdp-by-metro-area/

```python
gdp_per_capita = pd.DataFrame([60622.04, 64756.86,83596.75,61861.24,62255.63,66243.21,66819.75
                           ,0,58570.51,74689.26,0,0,57592.48,0,63006.38,93955.57,0,0,75973.11])
gdp_per_capita
```

```python
gdp_df = pd.concat([cities, gdp_per_capita], axis=1)
gdp_df.columns = ["city","gdp_per_capita"]
gdp_df = gdp_df.set_index("city")
gdp_df
```

```python
cab_df.head(5)
```

Investigate transaction counts per city. Cab table provides necessary informations.

```python
cab_use_df = cab_df.groupby(["city"]).agg({"transaction_id": "count", "km_travelled": "sum", "cost_of_trip": "sum", "price_charged": "sum"})
```

```python
cab_use_df["profit"] = - cab_use_df["cost_of_trip"] + cab_use_df["price_charged"]
```

```python
cab_use_df = pd.concat([cab_use_df, gdp_df], axis=1).drop(columns=["cost_of_trip", "price_charged"])
cab_use_df
```

Since data for some cities are  not available in dataset, gdp per capita with zero values will be dropped.

```python
cab_use_df = cab_use_df[cab_use_df["gdp_per_capita"] != 0]
```

```python
cab_use_df.corr()["gdp_per_capita"]
```

Unfortunately, gdp per capita and profit are not related.

```python
cab_use_df = cab_use_df.sort_values("profit", ascending=False)
cab_use_df.rename(columns={"transaction_id":"total_transactions"}, inplace=True)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 16))
sns.barplot(y = cab_use_df.index
            , x = cab_use_df["profit"]
            , ax=ax[0,0]).set(xlabel="Profit (x10^7)"
                              , ylabel="States")
sns.barplot(y = cab_use_df.index
            , x = cab_use_df["gdp_per_capita"]
            , ax=ax[0,1]).set(xlabel="GDP per Capita ($)"
                              , ylabel="States")
sns.barplot(y = cab_use_df.index
            , x = cab_use_df["km_travelled"]
            , ax=ax[1,0]).set(xlabel="Kilometer Travelled"
                              , ylabel="States")
sns.barplot(y = cab_use_df.index
            , x = cab_use_df["total_transactions"]
            , ax=ax[1,1]).set(xlabel="Total Transactions"
                              , ylabel="States")
```

New York shows the highest profit value. However, we also need to consider total number of transactions so that we understand profit per ride.

```python
cab_use_df["efficiency"] = cab_use_df["profit"].div(cab_use_df["total_transactions"].values, axis=0)
```

```python
cab_use_df
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.barplot(x=cab_use_df.index
            , y=cab_use_df["efficiency"] - cab_use_df["efficiency"].mean()
            , palette="crest").set(xlabel="Profit per Ride"
                                  , ylabel="States")
plt.xticks(rotation=90)
```

Graph is slided up through mean value of prfit per ride. The graph shows which state is more profitable.

## Hypothesis 2

Let us calculate profit amounts of each firm by comparing annual gasoline price.

```python
cab_df["profit"] = - cab_df["cost_of_trip"] + cab_df["price_charged"]
```

Smallest number of date of travel will represent 31/01/2016. So, smallest number will be extracted from all numbers and we are going to sum the resulted number to find the exact date.

```python
cab_df["date_of_travel"] -= cab_df["date_of_travel"].min()
```

```python
cab_df
```

```python
# os.environ['TZ'] = 'Europe/London'
```

```python
start_date = dt.datetime(2016,2,1).timestamp()
start_date
```

```python
one_day_timestamp = dt.datetime(2016,1,31).timestamp() - dt.datetime(2016,1,30).timestamp()
print("Timestamp value for 1 day:", one_day_timestamp)
```

```python
cab_df["travel_date"] = start_date
cab_df
```

```python
cab_df.dtypes
```

```python
cab_df["day_to_sum"] = cab_df["date_of_travel"]*one_day_timestamp
```

```python
# cab_df["ts"] = cab_df["travel-date"].values.astype(np.int64)
```

```python
cab_df
```

```python
# cab_df["travel-date-ts"] = cab_df["ts"] + cab_df["day-sum"]

cab_df["travel_date_ts"] = cab_df[["day_to_sum","travel_date"]].iloc[:,:].sum(axis=1)
```

```python
cab_df
```

```python
cab_df = cab_df.drop(columns=["travel_date", "day_to_sum"])
```

```python
cab_df
```

```python
cab_df["travel_date"] =  pd.to_datetime(cab_df["travel_date_ts"], unit="s")
```

```python
cab_df = cab_df.drop(columns=["travel_date_ts", "date_of_travel"])
```
